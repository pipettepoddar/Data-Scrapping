```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import requests
```

```python
url = 'https://www.goodreads.com/quotes/tag/motivational'
```

```python
html_good = requests.get(url).text
html_good
```

```python
soup = BeautifulSoup(html_good, "html.parser")
```

```python
def mquotes(soup):
    author_name = [i.text.strip() for i in soup.find_all('span',class_='authorOrTitle')]
    quotes = [list(i)[0].strip() for i in soup.find_all('div',class_='quoteText')]
    likes = [int(i.text.strip().split()[0]) for i in soup.find_all('a',class_='smallText')]
    tags = [i.text.split()[1:] for i in [i for i in soup.find_all('div',class_='greyText smallText left')]]
    return pd.DataFrame({"Authors":author_name,'Quotes':quotes,"Likes":likes,'Tags':tags})
```

```python
mquotes(soup)
```

```python
stud = 'https://www.goodreads.com/quotes/tag/motivational?page={}'
```

```python
tables=[]
from tqdm import tqdm
for i in tqdm(range(1,51)):
    url_all = stud.format(i)
    html_all = requests.get(url_all).text
    soup1 = BeautifulSoup(html_all,'html.parser')
    tables.append(mquotes(soup1))
```

```python
data = pd.concat(tables,axis=0)
data.head()
```

```python
data.to_csv('Motivational Quotes.csv')
```
